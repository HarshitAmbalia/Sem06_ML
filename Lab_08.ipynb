```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Load Data
from google.colab import files
import io
uploaded = files.upload()
datasets = pd.read_csv(io.BytesIO(uploaded['spam1.csv']))
datasets.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

dataset = datasets
dataset = pd.DataFrame(dataset)
dataset.columns = ["v2", "v1"]

nltk.download('stopwords')

print("dataset : ", dataset)
```

```python
corpus = []

#dataset[col][row]

for i in range(0, 513):
    text = re.sub('[^a-zA-Z]', '', dataset['v2'][i])
    text = text.lower()
    text = text.split()
    ps = PorterStemmer()
    text = ''.join(text)
    corpus.append(text)

print("corpus : ",corpus)
```

```python
# creating bag of words model
cv = CountVectorizer(max_features = 1500)

X = cv.fit_transform(corpus).toarray()
y =  dataset.iloc[:, 1].values
print("X : ", X)
print("y : ", y)
```

```python
# splitting the data set into training set and test set
import sklearn
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 98)
print("X_train : ",X_train)
print("X_test : ",X_test)
print("y_train : ",y_train)
print("y_test : ",y_test)
```

```python
# fitting naive bayes to the training set
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix

classifier = MultinomialNB();
classifier.fit(X_train, y_train)

# predicting test set results
y_pred = classifier.predict(X_test)
print("y_pred : ", y_pred)

# making the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("confusion_matrix : ", cm)
```

```python
#accuracy
from sklearn import metrics
ma = metrics.accuracy_score(y_test, y_pred)
print("Accuracy : ", ma)
```
